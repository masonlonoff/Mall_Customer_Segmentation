## EDA

**Importing** Neccessary **Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

**Reading in** the Dataset 

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

## High Level EDA

Finding the **Frequencies** of each **Numeric Column**

```python
df.hist(figsize=(12,8));
```

Checking for **Outliers**

```python
df.boxplot(figsize=(10,8));
```

There's only a single outlier and it's a reasonable value. The data point still contains valuable information so I won't be removing it.

Finding **Correlations** between each Column

```python
# Compute correlation matrix
corr_matrix = df.corr()

# Plot correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix', fontsize=15)
plt.xticks(rotation=45);
```

Examining the **Relationship** between the **Numeric Columns**

```python
# Select the relevant numerical columns for the scatter plot matrix
numerical_columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']

# Create a pairplot using seaborn
sns.pairplot(df[numerical_columns]);
```

## EDA on **CustomerID** Column

Finding **Unique** Values

```python
unique_values = df.CustomerID.unique()
print(unique_values)
```

Examining the **Value Counts**

```python
df.CustomerID.value_counts().sort_values()
```

## EDA on **Gender** Column

Examining the **Unique** Values in *Gender*

```python
unique_gender_values = df.Gender.unique()
print(unique_gender_values)
```

Checking out the **Value Counts**

```python
df.Gender.value_counts()
```

**Visualizing** the Counts of each Gender

```python
gender_counts = df.Gender.value_counts()

plt.bar(gender_counts.index, gender_counts.values)
plt.title('Number of Observations by Gender')
plt.xlabel('Gender')
plt.ylabel('Count');
```

Finding the **Unique Values**
